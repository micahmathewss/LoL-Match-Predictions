```python
import pandas as pd
import os

csv_directory = '/Users/micah.mathews/Documents/wager_ai/lol_data/data_extraction'
csv_files = [
    "2023_LoL_esports_match_data_from_OraclesElixir.csv",
    "2022_LoL_esports_match_data_from_OraclesElixir.csv",
    "2021_LoL_esports_match_data_from_OraclesElixir.csv",
    "2020_LoL_esports_match_data_from_OraclesElixir.csv",
    "2019_LoL_esports_match_data_from_OraclesElixir.csv",
    "2018_LoL_esports_match_data_from_OraclesElixir.csv",
    "2017_LoL_esports_match_data_from_OraclesElixir.csv",
    "2016_LoL_esports_match_data_from_OraclesElixir.csv",
    "2015_LoL_esports_match_data_from_OraclesElixir.csv"
]

# Create an empty DataFrame
combined_df = pd.DataFrame()

for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    try:
        df = pd.read_csv(file_path)
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"Appended: {csv_file}")
    except FileNotFoundError:
        print(f"File not found: {csv_file}")
    except Exception as e:
        print(f"Error while reading {csv_file}: {e}")

print(combined_df)
```

```python
df = combined_df
pd.set_option('display.max_columns', None)
df
```

```python
selected_columns = ['gameid', 'datacompleteness', 'url', 'league', 'year', 'split', 'playoffs', 'date', 'game', 'patch', 'side', 'position', 'playername', 'teamname', "result", 'champion']
df = df[selected_columns]
df
```

```python
df.loc[:, 'date'] = pd.to_datetime(df['date'])

df.sort_values(by='date', inplace=True)

grouped = df.groupby(['playername', 'champion'])

total_games = grouped['result'].cumcount()
total_wins = grouped['result'].cumsum()

win_rate = total_wins / total_games

df.loc[:, 'player_champ_winrate'] = win_rate

df



```

```python
grouped = df.groupby(['champion', 'patch'])

total_games = grouped['result'].cumcount()
total_wins = grouped['result'].cumsum()

win_rate = total_wins / total_games
win_rate[total_games == 0] = np.nan

df.loc[:, 'champion_patch_winrate'] = win_rate

df
```

```python
pivot = df.set_index(['gameid', 'position', 'side'])['champion']

pivot = pivot.reset_index()

side_map = {'Red': 'Blue', 'Blue': 'Red'}

pivot['opposite_side'] = pivot['side'].map(side_map)

df = pd.merge(df, pivot, left_on=['gameid', 'position', 'side'], 
              right_on=['gameid', 'position', 'opposite_side'], suffixes=('', '_y'))

df = df.drop(columns=['side_y', 'opposite_side'])

df = df.rename(columns={'champion_y': 'opposing_champion'})

df


```

```python
df['date'] = pd.to_datetime(df['date'])

df.sort_values(by='date', inplace=True)

grouped = df.groupby(['champion', 'opposing_champion'])

total_games = grouped['result'].cumcount()
total_wins = grouped['result'].cumsum()

win_rate = total_wins / total_games

df['counter_winrate'] = win_rate

df
```

```python
df = df[df['position'] != 'team']
df
```

```python
df = df.drop(['league', 'year', 'split', 'playoffs', 'opposing_champion'], axis=1)
df
```

```python
df.dropna(subset=['player_champ_winrate', 'champion_patch_winrate', 'counter_winrate'], inplace=True)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df
```

```python
avg_win_rate = df.groupby(['teamname', 'gameid'])[['player_champ_winrate', 'champion_patch_winrate', 'counter_winrate']].mean()

avg_win_rate.columns = ['avg_player_champ_winrate', 'avg_champion_patch_winrate', 'avg_counter_winrate']

df = df.merge(avg_win_rate, on=['teamname', 'gameid'])

df
```

```python
import pandas as pd

df = df.drop_duplicates(subset=['teamname', 'gameid'])

df


```

```python
df = df[['gameid', 'side','teamname', 'result', 'avg_player_champ_winrate', 'avg_champion_patch_winrate', 'avg_counter_winrate']]

df
```

```python
import pandas as pd

pivot = df.set_index(['gameid', 'side'])['avg_player_champ_winrate']

pivot = pivot.reset_index()

side_map = {'Red': 'Blue', 'Blue': 'Red'}

pivot['opposite_side'] = pivot['side'].map(side_map)

df = pd.merge(df, pivot, left_on=['gameid', 'side'],
              right_on=['gameid', 'opposite_side'], suffixes=('', '_y'))

df = df.drop(columns=['side_y', 'opposite_side'])

df = df.rename(columns={'avg_player_champ_winrate_y': 'opposing_avg_player_champ_winrate'})

df
```

```python
import pandas as pd

pivot = df.set_index(['gameid', 'side'])['avg_champion_patch_winrate']

pivot = pivot.reset_index()

side_map = {'Red': 'Blue', 'Blue': 'Red'}

pivot['opposite_side'] = pivot['side'].map(side_map)

df = pd.merge(df, pivot, left_on=['gameid', 'side'],
              right_on=['gameid', 'opposite_side'], suffixes=('', '_y'))

df = df.drop(columns=['side_y', 'opposite_side'])

df = df.rename(columns={'avg_champion_patch_winrate_y': 'opposing_avg_champion_patch_winrate'})

df
```

```python
pivot = df.set_index(['gameid', 'side'])['avg_counter_winrate']

pivot = pivot.reset_index()

side_map = {'Red': 'Blue', 'Blue': 'Red'}

pivot['opposite_side'] = pivot['side'].map(side_map)

df = pd.merge(df, pivot, left_on=['gameid', 'side'],
              right_on=['gameid', 'opposite_side'], suffixes=('', '_y'))

df = df.drop(columns=['side_y', 'opposite_side'])

df = df.rename(columns={'avg_counter_winrate_y': 'opposing_avg_counter_winrate'})

df
```

```python
df['difference_avg_player_champ_winrate'] = pd.NA
df['difference_avg_champion_patch_winrate'] = pd.NA
df['difference_avg_counter_winrate'] = pd.NA

df['difference_avg_player_champ_winrate'] = df['avg_player_champ_winrate'] - df['opposing_avg_player_champ_winrate']
df['difference_avg_champion_patch_winrate'] = df['avg_champion_patch_winrate'] - df['opposing_avg_champion_patch_winrate']
df['difference_avg_counter_winrate'] = df['avg_counter_winrate'] - df['opposing_avg_counter_winrate']

df
```

```python
df = df.drop(columns=['avg_player_champ_winrate', 'avg_champion_patch_winrate', 'avg_counter_winrate',
                      'opposing_avg_player_champ_winrate', 'opposing_avg_champion_patch_winrate',
                      'opposing_avg_counter_winrate'])
```
